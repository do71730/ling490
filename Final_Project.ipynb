```python
import string
import re
def remove_punctuation(text):
    text = re.sub('[^A-Za-z0-9\']+', ' ', text)
    text = re.sub('[^A-Za-z0-9 ]+', '', text)
    return text
```

```python
#look at the accourance of words in the corpus
from sklearn.feature_extraction.text import CountVectorizer
def countVectory(lyrics1,lyrics2,lyrics3,lyrics4):
  corpus = []
  lyrics1 = remove_punctuation(lyrics1)
  lyrics2 = remove_punctuation(lyrics2)
  lyrics3 = remove_punctuation(lyrics3)
  lyrics4 = remove_punctuation(lyrics4)
  corpus.append(lyrics1)
  corpus.append(lyrics2)
  corpus.append(lyrics3)
  corpus.append(lyrics4)
  vectorizer = CountVectorizer()
  X = vectorizer.fit_transform(corpus)
  df_countvect = pd.DataFrame(data = X.toarray(),index = ['lyrics1','lyrics2', 'lyrics3','lyrics4'],columns =vectorizer.get_feature_names_out())
  return df_countvect
```

```python
def bigram(text):
  text = remove_punctuation(text)
  tokens = nltk.word_tokenize(text)
  bigrams = nltk.bigrams(tokens)
  frequence = nltk.FreqDist(bigrams)
  for key,value in frequence.items():
      print(key,value)
```
